```python
filtered_paths = []
id_set = set()

for path in item_color_data['image_name']:
    parts = path.split('/')
    id_ = parts[3] # the id_ value is the 4th part of the path
    filename = parts[-1] # the filename is the last part of the path
    if id_ not in id_set and filename.endswith('front.jpg'):
        filtered_paths.append(path)
        id_set.add(id_)
print(len(filtered_paths))
print(filtered_paths)
```

```python
original_path= filtered_paths[0]
parts = original_path.split('/')
new_path = '/'.join(parts[3:4]) + '.jpg'
print(new_path)
```
